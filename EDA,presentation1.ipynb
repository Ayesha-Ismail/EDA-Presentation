```python
 #Exploratory Data Analysis (EDA)
 #1. Visualize Distributions
#Use Seaborn and Matplotlib for visualizations.
import seaborn as sns
import matplotlib.pyplot as plt

# Pairplot for visualizing the relationships
sns.pairplot(df, hue='species')
plt.title('Pairplot of Iris Dataset')
plt.show()
```

```python
# prompt: Correlation Matrix
# #Check for correlations between features.
corr_matrix = df.drop(columns=['species']).corr() # Drop the 'species' column before calculating correlation
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Iris Dataset')
plt.show()
```

```python
# prompt: insights of correlation

# Based on the correlation matrix, we can see the following insights:

# Strong positive correlation:
# - Sepal length and sepal width: 0.72, indicating that as sepal length increases, sepal width tends to increase as well.
# - Petal length and petal width: 0.96, indicating a very strong positive relationship.
#   Larger petal length generally corresponds to larger petal width.

# Moderate positive correlation:
# - Petal length and sepal length: 0.87, indicating a strong relationship between petal and sepal length.
# - Petal width and sepal length: 0.82, showing a good relationship between petal width and sepal length.
# - Petal width and sepal width: 0.42, suggesting a somewhat weaker positive relationship between petal width and sepal width.


# These correlations suggest that the features are somewhat inter-related. We can explore these relationships further to understand how different features are linked to each other, which can contribute to identifying patterns within the dataset.


# Note that correlation does not imply causation. It simply identifies a statistical relationship between variables.
```

```python
# prompt: . Target Variable Analysis
# Visualize the distribution of species.

plt.figure(figsize=(8, 6))
sns.countplot(x='species', data=df)
plt.title('Distribution of Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
# prompt: Create an Insight Dashboard

import plotly.express as px

# Create a scatter plot for sepal length vs. sepal width
fig_scatter = px.scatter(df, x="sepal_length", y="sepal_width", color="species", title="Sepal Length vs. Sepal Width by Species")
fig_scatter.show()

```

```python
# Create a histogram for sepal length
fig_hist = px.histogram(df, x="sepal_length", color="species", title="Distribution of Sepal Length by Species")
fig_hist.show()
```

```python
# prompt: Boxplot to see distribution of petal length by species

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal_length', data=df)
plt.title('Distribution of Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```
